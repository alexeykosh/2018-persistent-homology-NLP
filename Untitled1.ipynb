```python
import nltk
import matplotlib.pyplot as plt
from gensim.models import word2vec, KeyedVectors
from persistent.persistent import Persistent
import numpy as np
from itertools import islice
from random import randint, sample
import wikipedia
import dionysus as dn
from sklearn.manifold import TSNE
```

```python
def get_texts_for_lang(lang, n=10):
    wikipedia.set_lang(lang)
    wiki_content = []
    pages = wikipedia.random(n)
    for page_name in pages:
        try:
            page = wikipedia.page(page_name)
        except wikipedia.exceptions.WikipediaException:
            print('Skipping page {}'.format(page_name))
            continue

        wiki_content.append('{}\n{}'.format(page.title, page.content.replace('==', '')))

    return wiki_content
```

### Functions:

```python
flatten = lambda l: [item for sublist in l for item in sublist]

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()
```

```python
def random_chunk(li, min_chunk=1, max_chunk=3):
    it = iter(li)
    while True:
        nxt = list(islice(it,randint(min_chunk,max_chunk)))
        if nxt:
            yield nxt
        else:
            break
```

```python
def scramble(text):
    text = text.replace('[^\w\s]','')
    text = [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text.lower())] 
    text_f = sample(text, len(text))
    
    return text_f, text
```

```python
scramble('You can look at leipzig.md for a full example of how to do anything that the The Leipzig Glossing Rules allow for.')
```

### Experiment

Pre-trained model:

```python
model = KeyedVectors.load_word2vec_format('persistent/GoogleNews-vectors'
                                              '-negative300.bin', binary=True)
```

```python
texts = get_texts_for_lang('en', 100)
```

```python
res = []

for text in texts:
    a = Persistent(model=model, split_sent=scramble(text)[1])
    b = Persistent(model=model, split_sent=scramble(text)[0])
    
    res.append((len(text), dn.wasserstein_distance(a.persistent()[0], b.persistent()[0])))
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(y=[i[1] for i in res], x=[i[0] for i in res])
plt.show()
```

### Tests:

```python
def tsne_plot(tokens):
    """Creates and TSNE model and plots it"""    
    perplexitys = [5, 30, 50, 100]
    
    position = 221
    
    plt.figure(1, figsize=(10, 10)) 

    
    for i in range(len(perplexitys)):
        tsne_model = TSNE(perplexity=perplexitys[i], n_components=2, init='pca', n_iter=5000, random_state=23)
        new_values = tsne_model.fit_transform(tokens)

        x = []
        y = []
        for value in new_values:
            x.append(value[0])
            y.append(value[1])
        
        plt.subplot(position)
        plt.title('Perplexity = {}'.format(perplexitys[i]))
        
        for i in range(len(x)):
            plt.scatter(x[i],y[i])

        position += 1
        
    plt.show()
```

```python
def create_tokens(sent):
    tokens = []
    
    for w in sent:
        try:
            tokens.append(model[w])
        except:
            pass
    
    return tokens
```

```python
tsne_plot(create_tokens(scramble(texts[21])[0]))
```
